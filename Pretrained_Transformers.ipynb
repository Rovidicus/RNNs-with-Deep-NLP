```python
from transformers import pipeline
```

```python
text = "I liked the food but the service was poor and we had to wait a very long time."
```

```python
# Choose a pre-trained model suitable for your task
# Note: this is the default module used for sentiment-analysis if no model specified
model = "distilbert-base-uncased-finetuned-sst-2-english"
```

```python
# Sentiment analysis
classifier = pipeline("sentiment-analysis", model=model)
classifier(text)
```

```python
## Zero-shot classification
​​# Note: this is the default module used for zero-shot classifications if no model is specified.
model = 'roberta-large-mnli'
classifier = pipeline("zero-shot-classification", model=model)
classifier(text,
    candidate_labels=["restaurant", "movie theater", "hospital"])
```
